```python
# 90점 이상이면 'A'
# 80점 이상 90점 미만이면 'B'
# 70점 이상 80점 미만이면 'C'
# 60점 이상 70점 미만이면 'D'
# 아니면 'F'
```

```python

score = 65
if score >= 90:
    result= 'A'
elif score >= 80:
    result= 'B'
elif score >= 70:
    result= 'C'
elif score >= 60:
    result= 'D'   
else:
    result = 'F'
    

result
print(F'학점은 {result}입니다')
```

```python
# 수학점수와 영어 점수가 둘 다 90점 이상이면 '참 잘했어요'
# 수학점수만 90점 이상이면 '좀 더 노력하세요' 
# 그렇지않으면 '많이 노력하세요' 


math = 100
eng = 80

if math >= 90:
    if eng >= 90:
        print('참 잘했어요')
    else:
        print('좀 더노력하세요')
else:
    print('많이 노력하세요')

```

```python
# while 반복문을 이용해서 1부터 10까지 더하되 합이 20이 넘는 순간의 정수값을  출력하고 while문이 종료되는 코드 작성
# 정수: ?, 합: > 20



sum = 0
i = 0

while True:
    i += 1
    sum += i 
    if sum >= 20:
        print(f'정수: {i}, 합: {sum}')
        break
```

```python
import random
com = random.randint(1,100)

print(com)

while True:
    user = int(input('원하는 숫자'))
    if user == com:
        print('정답입니다!')
        break
    elif user > com:
        print('너무커요')
    else:
        print('너무작아요')
```
